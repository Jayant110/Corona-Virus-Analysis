# Covid - 19 Analysis

#### Data Cleaning & Transformation

```python
import numpy as np
import pandas as pd
from IPython.display import display
```

```python
# importing datsets
raw_data_confirmed = pd.read_csv(r"G:\08 - Github Projects\EDA & Machine Learning\Corona Virus\Raw\Covid19_confirmed_global.csv")
raw_data_deaths = pd.read_csv(r"G:\08 - Github Projects\EDA & Machine Learning\Corona Virus\Raw\Covid19_deaths_global.csv")
raw_data_recovered = pd.read_csv(r"G:\08 - Github Projects\EDA & Machine Learning\Corona Virus\Raw\Covid19_recovered_global.csv")
```

```python
display(f"Confirmed dataset - {raw_data_confirmed.shape}")
display(f"Death dataset - {raw_data_deaths.shape}")
display(f"Global dataset - {raw_data_recovered.shape}")
display(raw_data_confirmed.head())
```

```python
# unpivoting data
raw_data_confirmed = pd.melt(raw_data_confirmed, id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
                              var_name=['Date'])
raw_data_deaths = pd.melt(raw_data_deaths, id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
                           var_name=['Date'])
raw_data_recovered = pd.melt(raw_data_recovered, id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
                              var_name=['Date'])

display(f"Confirmed dataset - {raw_data_confirmed.shape}")
display(f"Death dataset - {raw_data_deaths.shape}")
display(f"Global dataset - {raw_data_recovered.shape}")
        
display(raw_data_confirmed.head())
```

```python
# we have to convert type of "date" to date
raw_data_confirmed['Date'] = pd.to_datetime(raw_data_confirmed['Date'])
raw_data_deaths['Date'] = pd.to_datetime(raw_data_deaths['Date'])
raw_data_recovered['Date'] = pd.to_datetime(raw_data_recovered['Date'])
```

```python
# renaming columns
raw_data_confirmed.rename(columns={'value' : 'Confirmed'}, inplace=True)
raw_data_deaths.rename(columns={'value' : 'Deaths'}, inplace=True)
raw_data_recovered.rename(columns={'value' : 'Recovered'}, inplace=True)
```

```python
display(raw_data_confirmed.head())
display(raw_data_deaths.head())
display(raw_data_recovered.head())
```

```python
# checking for null values
display(raw_data_confirmed.isnull().sum())
display(raw_data_deaths.isnull().sum())
display(raw_data_recovered.isnull().sum())
```

```python
# handling null values
raw_data_confirmed['Province/State'].fillna(raw_data_confirmed['Country/Region'], inplace=True)
raw_data_deaths['Province/State'].fillna(raw_data_deaths['Country/Region'], inplace=True)
raw_data_recovered['Province/State'].fillna(raw_data_recovered['Country/Region'], inplace=True)
```

```python
# joining
full_join = raw_data_confirmed.merge(raw_data_deaths[['Province/State','Country/Region','Date','Deaths']], 
                                      how = 'left', 
                                      left_on = ['Province/State','Country/Region','Date'], 
                                      right_on = ['Province/State', 'Country/Region','Date'])
full_join.head()
```

```python
full_join = full_join.merge(raw_data_recovered[['Province/State','Country/Region','Date','Recovered']], 
                                     how = 'left', 
                                     left_on = ['Province/State','Country/Region','Date'], 
                                     right_on = ['Province/State', 'Country/Region','Date'])
full_join.head()
```

```python
# adding "Month-year" column
full_join['Month-Year'] = full_join['Date'].dt.strftime('%b-%Y')

# adding Active cases
full_join["Active Cases"] = full_join["Confirmed"] - (full_join["Recovered"] - full_join["Deaths"])

# rearranging columns
full_join = full_join[['Province/State', 'Country/Region', 'Lat', 'Long', 'Date', 'Month-Year',
                       'Confirmed', 'Deaths', 'Recovered', 'Active Cases']]
```

```python
full_join
```

##### Calculating Daily values for Confirmed, Deaths, Recovered & Active Cases

```python
# sorting date
full_join.sort_values('Date', inplace=True)

# grouping by
grouped = full_join.groupby(['Country/Region', 'Province/State'])

# the diff function calculates the difference between the current row and the previous row
daily_cases = grouped[['Confirmed', 'Deaths', 'Recovered', 'Active Cases']].diff().fillna(0)

# joiining back to the dataframe
full_join = full_join.join(daily_cases, rsuffix='_daily')
```

```python
# handling negative values
def negtozero(x):
    
    if x < 0:
        return 0
    else:
        return x
```

```python
# applying
full_join["Confirmed_daily"] = full_join["Confirmed_daily"].apply(negtozero)
full_join["Deaths_daily"] = full_join["Deaths_daily"].apply(negtozero)
full_join["Recovered_daily"] = full_join["Recovered_daily"].apply(negtozero)
full_join["Active Cases_daily"] = full_join["Active Cases_daily"].apply(negtozero)
```
